```python
# =========================================================
# DATA VISUALIZATION II - TITANIC DATASET
# BOX PLOT: AGE vs GENDER + SURVIVAL STATUS
# =========================================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')

print("Dataset Loaded\n")
print(df.head())

# =========================================================
# Data Cleaning (Important)
# =========================================================
df = df.dropna(subset=['age'])   # remove missing age values

# =========================================================
# 1. BOXPLOT: AGE vs SEX with SURVIVAL INFO
# =========================================================

plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=df,
    palette='Set2'
)

plt.title("Age Distribution by Gender and Survival Status", fontsize=12)
plt.xlabel("Gender (Sex)")
plt.ylabel("Age")
plt.legend(title="Survived (0=No, 1=Yes)")

plt.tight_layout()
plt.show()

# =========================================================
# 2. OBSERVATIONS (PRINTED FOR LAB RECORD)
# =========================================================

print("\n================ OBSERVATIONS ================\n")

print("1. Female passengers have significantly higher survival rates compared to males.")
print("2. Most non-survivors are males in the adult age group (20–50 years).")
print("3. Children (lower age range) show higher chances of survival, supporting 'women and children first'.")
print("4. The median age of survivors is slightly lower than that of non-survivors.")
print("5. Age distribution is wider for males, indicating more variation in survival outcomes.")
print("6. Presence of outliers shows a few very young and very old passengers.")

print("\n================================================")
```
