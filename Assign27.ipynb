```python
# ============================================
# 📊 DATA VISUALIZATION - GDP DATASET
# ============================================

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("gdp.csv")

# ============================================
# a) BAR GRAPH: Belgium GDP (2006–2026)
# ============================================

belgium = df[(df['country'] == 'Belgium') & 
             (df['year'] >= 2006) & (df['year'] <= 2026)]

plt.figure(figsize=(8,5))
plt.bar(belgium['year'], belgium['gdpPercap'])
plt.title("Belgium GDP (2006–2026)")
plt.xlabel("Year")
plt.ylabel("GDP per Capita")
plt.xticks(rotation=45)
plt.show()


# ============================================
# b) PIE CHART: GDP (Year 2010)
# Countries: India, Nepal, Romania, South Asia, Singapore
# ============================================

countries_2010 = df[
    (df['year'] == 2010) & 
    (df['country'].isin(['India', 'Nepal', 'Romania', 'Singapore']))
]

plt.figure(figsize=(6,6))
plt.pie(countries_2010['gdpPercap'],
        labels=countries_2010['country'],
        autopct='%1.1f%%')
plt.title("GDP Distribution (2010)")
plt.show()


# ============================================
# c) LINE GRAPH: Bhutan vs Costa Rica
# ============================================

bhutan = df[df['country'] == 'Bhutan']
costa_rica = df[df['country'] == 'Costa Rica']

plt.figure(figsize=(8,5))
plt.plot(bhutan['year'], bhutan['gdpPercap'], label='Bhutan', marker='o')
plt.plot(costa_rica['year'], costa_rica['gdpPercap'], label='Costa Rica', marker='o')

plt.title("GDP Comparison: Bhutan vs Costa Rica")
plt.xlabel("Year")
plt.ylabel("GDP per Capita")
plt.legend()
plt.grid()
plt.show()


# ============================================
# d) SCATTER PLOT: 2004–2006
# Countries: Mexico, Algeria, Fiji, Estonia
# ============================================

scatter_data = df[
    (df['year'].between(2004, 2006)) &
    (df['country'].isin(['Mexico', 'Algeria', 'Fiji', 'Estonia']))
]

plt.figure(figsize=(8,5))

for country in scatter_data['country'].unique():
    temp = scatter_data[scatter_data['country'] == country]
    plt.scatter(temp['year'], temp['gdpPercap'], label=country)

plt.title("GDP Scatter Plot (2004–2006)")
plt.xlabel("Year")
plt.ylabel("GDP per Capita")
plt.legend()
plt.grid()
plt.show()
```
