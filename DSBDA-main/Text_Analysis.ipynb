```python
# Dushyant Krishna Sharma
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

text = """Text analytics is the process of deriving meaningful information from text.
It involves tokenization, stop word removal, stemming, and lemmatization."""
# Tockenisation
tokens = word_tokenize(text)
print(tokens)
```

```python
# Pos tagging
nltk.download('averaged_perceptron_tagger')

pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
# Stop Words
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

```python
# Stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print(stemmed_words)
```

```python
# Lemmatisation
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemmatized_words)
```

```python
documents = [
    "Text analytics is useful",
    "Text mining and analytics are related",
    "Analytics derives insights from text data"
]
```

```python
# TF (Term Frequency)
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
# TF-IDF(Inverse document frequency)
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(documents)

print(tfidf.get_feature_names_out())
print(X_tfidf.toarray())
```
