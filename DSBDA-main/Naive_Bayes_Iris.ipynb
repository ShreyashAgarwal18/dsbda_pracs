```python
#Dushyant krishna sharma
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
# Load the dataset
dataset = pd.read_csv('Iris.csv')

# Select input features (X)
X = dataset[['SepalLengthCm', 'SepalWidthCm',
             'PetalLengthCm', 'PetalWidthCm']].values

# Select output label (y)
y = dataset['Species'].values

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# Get unique species names
class_names = np.unique(y)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=class_names)

# Convert to DataFrame for clear display
cm_df = pd.DataFrame(cm, index=class_names, columns=class_names)

print("Confusion Matrix with Species Names:")
print(cm_df)
```

```python
plt.figure(figsize=(7, 5))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Species')
plt.ylabel('Actual Species')
plt.title('Confusion Matrix – Naïve Bayes (Iris Dataset)')
plt.show()
```
