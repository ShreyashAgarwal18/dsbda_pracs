```python
# ===========================================
# DESCRIPTIVE STATISTICS (IRIS DATASET)
# ===========================================

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

print("\nIRIS DATASET HEAD:")
print(df.head())

# Categorical column
categorical_col = "species"

# Numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

# Grouping + Summary statistics
grouped_stats = df.groupby(categorical_col)[numeric_cols].agg(
    ['mean', 'median', 'min', 'max', 'std']
)

print("\nSUMMARY STATISTICS (GROUPED BY SPECIES):")
print(grouped_stats)

# List of values for one numeric feature (sepal length)
print("\nSEPAL LENGTH VALUES PER SPECIES:")
for s, values in df.groupby('species')['sepal length (cm)']:
    print("Species", s, ":", list(values))
```

```python
import pandas as pd
import numpy as np

# Load dataset (normal CSV format)
df = pd.read_csv("winequality-red.csv")

print("\nHEAD:")
print(df.head())

# Categorical column
categorical_col = "quality"

# Numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

# Summary statistics
grouped_stats = df.groupby(categorical_col)[numeric_cols].agg(
    ['mean', 'median', 'min', 'max', 'std']
)

print("\nSUMMARY STATISTICS:")
print(grouped_stats)

# List of values (alcohol per quality)
print("\nALCOHOL VALUES PER QUALITY:")
for q, values in df.groupby('quality')['alcohol']:
    print("Quality", q, ":", list(values))
```
