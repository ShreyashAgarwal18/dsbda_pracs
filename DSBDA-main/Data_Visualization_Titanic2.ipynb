```python
#Dushyant Krishna Sharma
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('titanic.csv')
print(titanic_data.columns)
print(titanic_data)
```

```python
titanic_data['Survived'] = titanic_data['Survived'].map({0: 'Not Survived', 1: 'Survived'})
sns.set(style="whitegrid")
print(titanic_data)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic_data, palette='Set2')
plt.title('Age Distribution by Gender and Survival Status', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.countplot(x='Sex', hue='Survived', data=titanic_data, palette='Set2')

plt.title('Count of Men and Women by Survival Status', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```
