```python
# ==============================
# Import Libraries
# ==============================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================
# Load Dataset
# ==============================
df = sns.load_dataset('titanic')

# ==============================
# Basic Info
# ==============================
print(df.head())
print(df.info())
print(df.describe())

# ==============================
# 1. Survival Count
# ==============================
sns.countplot(x='survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()

# ==============================
# 2. Survival based on Gender
# ==============================
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival based on Gender")
plt.show()

# ==============================
# 3. Passenger Class Distribution
# ==============================
sns.countplot(x='pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()

# ==============================
# 4. Survival based on Class
# ==============================
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival based on Passenger Class")
plt.show()

# ==============================
# 5. Age Distribution
# ==============================
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# ==============================
# 6. Fare Distribution (MAIN QUESTION)
# ==============================
plt.figure(figsize=(8,5))
sns.histplot(df['fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()

# ==============================
# 7. Boxplot for Fare (Outliers)
# ==============================
sns.boxplot(x=df['fare'])
plt.title("Fare Boxplot (Outliers)")
plt.show()

# ==============================
# 8. Age vs Fare (Scatter Plot)
# ==============================
sns.scatterplot(x='age', y='fare', data=df)
plt.title("Age vs Fare")
plt.show()

# ==============================
# 9. Survival vs Age (Boxplot)
# ==============================
sns.boxplot(x='survived', y='age', data=df)
plt.title("Survival vs Age")
plt.show()

# ==============================
# 10. Correlation Heatmap
# ==============================
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
