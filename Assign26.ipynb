```python
# =========================================================
# DATA VISUALIZATION - DEATH RATE DATASET
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# 🔹 STEP 1: LOAD DATASET
# =========================================================

# If file is in same folder
df = pd.read_csv("death_rate.csv")

print("\n========== DATASET ==========\n")
print(df.head())

# =========================================================
# 🔹 STEP 2: BASIC INFO
# =========================================================

print("\n========== DATASET INFO ==========\n")
print(df.info())

# =========================================================
# 🔹 PART (a): VISUALIZE DATA USING DIFFERENT GRAPHS
# =========================================================

# 1. BAR GRAPH: Total deaths per country
country_deaths = df.groupby("Country")["Number_of_Deaths"].sum()

plt.figure()
country_deaths.plot(kind='bar')
plt.title("Total Deaths by Country")
plt.xlabel("Country")
plt.ylabel("Number of Deaths")
plt.show()


# 2. LINE GRAPH: Deaths over years
year_deaths = df.groupby("Year")["Number_of_Deaths"].sum()

plt.figure()
year_deaths.plot(marker='o')
plt.title("Deaths Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.show()


# 3. PIE CHART: Deaths by Gender
gender_deaths = df.groupby("Gender")["Number_of_Deaths"].sum()

plt.figure()
gender_deaths.plot(kind='pie', autopct='%1.1f%%')
plt.title("Death Distribution by Gender")
plt.ylabel("")
plt.show()


# 4. HISTOGRAM: Mortality Rate Distribution
plt.figure()
plt.hist(df["Mortality_Rate_per_1000"])
plt.title("Mortality Rate Distribution")
plt.xlabel("Mortality Rate")
plt.ylabel("Frequency")
plt.show()


# =========================================================
# 🔹 PART (b): YEAR WITH MAXIMUM ACCIDENTS
# =========================================================

# Filter only Road Accidents
accidents = df[df["Cause_of_Death"] == "Road Accidents"]

year_accidents = accidents.groupby("Year")["Number_of_Deaths"].sum()

max_year = year_accidents.idxmax()
max_value = year_accidents.max()

print("\n========== MAX ACCIDENT YEAR ==========\n")
print(f"Year with maximum road accidents: {max_year}")
print(f"Number of deaths: {max_value}")


# =========================================================
# 🔹 PART (c): STATE/COUNTRY-WISE DEATH COUNT
# =========================================================

# (Dataset has Country instead of State)
country_deaths = df.groupby("Country")["Number_of_Deaths"].sum().sort_values()

plt.figure()
country_deaths.plot(kind='barh')
plt.title("Country-wise Death Count")
plt.xlabel("Number of Deaths")
plt.ylabel("Country")
plt.show()
```
