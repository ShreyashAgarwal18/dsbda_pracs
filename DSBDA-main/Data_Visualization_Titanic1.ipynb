```python
#Dushyant krishna sharma
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Fare', hue='Sex', bins=30, kde=True, palette='Set1')
plt.title('Comparison of Ticket Fare Distribution by Sex')
plt.xlabel('Fare')
plt.ylabel('Number of passengers')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Fare', hue='Pclass', bins=30, kde=True, palette='Set1')
plt.title('Ticket Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
df['Deck'] = df['Cabin'].apply(lambda x: x[0] if x != 'Unknown' else 'Unknown')
df_hist = df[df['Fare'] < 500]  

plt.figure(figsize=(10,6))
sns.histplot(data=df_hist, x='Fare', hue='Deck', bins=30, kde=True, palette='Set3')
plt.title('Ticket Fare Distribution by Deck')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python

df_clean = df.dropna(subset=['Age', 'Fare']) 
df_clean = df_clean[df_clean['Fare'] < 500]  


df_clean['AgeGroup'] = pd.cut(df_clean['Age'], bins=[0,12,18,30,50,80], labels=['Child','Teen','Young Adult','Adult','Senior'])


plt.figure(figsize=(10,6))
sns.histplot(data=df_clean, x='Fare', hue='AgeGroup', bins=30, kde=True, palette='Set2')
plt.title('Ticket Fare Distribution by Age Group')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
df_clean = df[df['Fare'] < 500]

plt.figure(figsize=(10,6))
sns.histplot(data=df_clean, x='Fare', hue='SibSp', bins=30, kde=True, palette='Set2')
plt.title('Ticket Fare Distribution by Number of Siblings/Spouses (SibSp)')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Fare', hue='Survived', bins=30, kde=True, palette='Set2')
plt.title('Comparison of Ticket Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Number of passengers')
plt.show()
```
